```python
list_pronouns = re.split(',|y', pronouns)

# Remove leading and trailing spaces
list_pronouns = [pronoun.strip() for pronoun in list_pronouns]

# Remove punctuation
list_pronouns = [pronoun.translate(str.maketrans('', '', string.punctuation)) for pronoun in list_pronouns]
```

```python
list_phrases = phrases.split('\n')

# Filter out empty items
list_phrases = [phrase for phrase in list_phrases if phrase]

list_phrases
```

```python
new_list = []

for f in list_phrases:
    new_words = []
    replaced_pronouns = []
    for w in f.split(' '):
        for p in list_pronouns:
            if w.lower() == p:
                replaced_pronouns.append(p)
                w = '_____'
        new_words.append(w)
    new_phrase = ' '.join(new_words)
    new_list.append((new_phrase, replaced_pronouns))
```

```python
for phrase, pronouns in new_list:
    for _ in range(3):  # give the user 3 attempts
        user_input = input(f"Complete the phrase: {phrase}\n")
        if user_input.lower() == "exit":
            print("Exiting the game.")
            break
        elif user_input.lower() in pronouns:
            print("Correct!")
            break
        else:
            print("Incorrect, try again.")
    else:
        print(f"You didn't guess the correct pronoun. The correct answer was: {pronouns}")
    if user_input.lower() == "exit":
        break
```
